# Scatter

Scatter can show up to 4 dimensions of data controling the (x, y, shape, color) of the markers based on your input.  In additionally you can style the shape unifromly across a single scatter call.

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("http://bit.ly/tscv17")
```

You can use `df.head()` to take a quick peak at the data.  This is helpful to remind ourselves what the column names and datatypes are!

```python
df.head()
```

To start with, lets plot the passanger age against the fare paid:

```python
fig, ax = plt.subplots()
ax.scatter(df['Age'], df['Fare'])
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
```

```python
fig, ax = plt.subplots()
ax.scatter(df['Age'], df['Fare'], c=df['Survived'])
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
```

```python
fig, ax = plt.subplots()
ax.scatter(df['Age'], df['Fare'], c=df['Survived'], alpha=.5, linewidth=0)
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
```

```python
fig, ax = plt.subplots()
ax.scatter(df['Age'], df['Fare'], c=df['Survived'], alpha=.5, linewidth=0, s=25/df['Pclass'])
ax.set_xlabel('Age')
ax.set_ylabel('Fare')
```

```python
fig, ax = plt.subplots()
x, y, c, s = 'Age', 'Fare', 'Survived', 'Pclass'

ax.scatter(x, y, c=c, s=25/df[s], alpha=.5, linewidth=0, data=df)
ax.set_xlabel(x)
ax.set_ylabel(y)
```

```python
fig, ax = plt.subplots()
x, y, c, s = 'Age', 'Fare', 'Survived', 'Pclass'

ax.scatter(x, y, c=c, alpha=.5, linewidth=0, marker='s', data=df[df['Sex'] =='male'])
ax.scatter(x, y, c=c, alpha=.5, linewidth=0, marker='^', data=df[df['Sex'] =='female'])
ax.set_xlabel(x)
ax.set_ylabel(y)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
x, y, c, s = 'Age', 'Fare', 'Survived', 'Pclass'

ax1.scatter(x, y, c=c, alpha=.5, linewidth=0, marker='s', data=df[df['Sex'] =='male'])
ax2.scatter(x, y, c=c, alpha=.5, linewidth=0, marker='^', data=df[df['Sex'] =='female'])
ax1.set(ylabel=y, title='male')
ax2.set(xlabel=x, ylabel=y, title='female')
```

```python
fig, ax = plt.subplots()
x, y, c, s = 'Age', 'Fare', 'Survived', 'Pclass'
markers = {'male': 's', 'female': '^'}
for gender, gdf in df.groupby('Sex'):
    ax.scatter(x, y, c=c, alpha=.5, linewidth=0, marker=markers[gender], data=gdf, label=gender)

ax.set_xlabel(x)
ax.set_ylabel(y)
ax.legend()
```
